## Label Encoders

```python
# leave and college using LabelEncoder
leave_encoder = LabelEncoder()
df["label"] = leave_encoder.fit_transform(df["LEAVE"]) 
# as an alternative using OrdinalEncoder to make the mapping explicit (STAY=0, LEAVE=1)
leave_encoder = OrdinalEncoder(categories=[["STAY","LEAVE"]],dtype=int)
df["label"] = leave_encoder.fit_transform(df[["LEAVE"]])
col_encoder = LabelEncoder()
df["COLLEGE_NUM"] = col_encoder.fit_transform(df["COLLEGE"])
# as an alternative using OrdinalEncoder to make the mapping explicit (zero=0, one=1)
col_encoder = OrdinalEncoder(categories=[["zero","one"]],dtype=int)
df["COLLEGE_NUM"] = col_encoder.fit_transform(df[["COLLEGE"]])
```

## Ordinal Encoders

```python
# Ordinal encoding for the ordinal attributes (considering change of plan, satisfaction, usage level)
change_encoder = OrdinalEncoder(categories=[['no','never_thought','perhaps', 'considering', 'actively_looking_into_it' ]], dtype=int)
df["CONSIDER_CHANGE_OF_PLAN_NUM"] = change_encoder.fit_transform(df[["CONSIDERING_CHANGE_OF_PLAN"]])    
sat_encoder = OrdinalEncoder(categories=[["very_unsat",'unsat','avg','sat','very_sat']], dtype=int)
df["REPORTED_SATISFACTION_NUM"] = sat_encoder.fit_transform(df[["REPORTED_SATISFACTION"]])
usage_encoder = OrdinalEncoder(categories=[['very_little','little','avg', 'high', 'very_high']], dtype=int)
df["REPORTED_USAGE_LEVEL_NUM"] = usage_encoder.fit_transform(df[["REPORTED_USAGE_LEVEL"]])
df
```
