# Sales Data Binning

```python
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
```

## Read the input file

```python
df = pd.read_csv("../data/sales_for_data_cleaning.csv",delimiter=";")
display(df)
```

Lower division

Filter lines with missing values

```python
print(len(df))
df_without_na = df.dropna()
print(len(df_without_na))
```

### Binning of Salary

```python
max_salary = round(df["salary"].max())
min_salary = round(df["salary"].min())
splits = [i for i in range(min_salary,max_salary,round((max_salary-min_salary)/10))]
print(splits)
labels = [f"{splits[i]} - {splits[i+1]}" for i in range(len(splits)-1)]
# Create a bin column
df['salary_bins'] = pd.cut(df['salary'], bins=splits, labels=labels, include_lowest=True)
display(df)
```

Discretize with scikit-learn

```python
equal_width_binning = KBinsDiscretizer(
    n_bins=10, encode="ordinal", strategy="uniform"
)
equal_hight_binning = KBinsDiscretizer(
    n_bins=10, encode="ordinal", strategy="quantile", quantile_method='averaged_inverted_cdf'
)
df["salary_bins_equal_width"] = equal_width_binning.fit(df[["salary"]]).transform(df[["salary"]])
df["salary_bins_equal_height"] = equal_hight_binning.fit(df[["salary"]]).transform(df[["salary"]])
df
```
