# IDA 2016 Correct Missing Value Coding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer
```

## Read the input file

```python
df = pd.read_csv("../data/ida_2016_training_set_update.csv", header=0, delimiter=",")
df 
```

## Replace na with None

```python
print(len(df))
print(len(df.dropna())) 
df_with_corr_null = df.replace(to_replace='na', value=None)
print(len(df_with_corr_null.dropna())) 
display(df_with_corr_null.dropna())
```

## Calculation of Null Values per Column

```python
Dict_Null = {col:len(df_with_corr_null[df_with_corr_null[col].isna()]) for col in df_with_corr_null.columns}
display(Dict_Null)
```
