# Sales Data Cleaning 

```python
import pandas as pd
```

## Read the input file

```python
df = pd.read_csv("../data/sales_for_data_cleaning.csv",delimiter=";")
display(df)
```

Lower division

```python
# df["division_lowered"] = df["division"].str.lower()
df["division_lowered"] = df["division"].str.casefold()
display(df)
```

Filter lines with missing values

```python
print(len(df))
df_without_na = df.dropna()
print(len(df_without_na))
display(df_without_na)
```

Fill missing values with avg

```python
# calculate which columns have null values
for curr_col in df.columns : 
   print(curr_col, " Null values ",df[curr_col].isnull().sum())#
# only training level and work experience have Null values
# calc avg for Training Level and 
avg_training_level = df["training level"].mean()
print (avg_training_level)
avg_work_experience = df["work experience"].mean()
print (avg_work_experience)
df_with_filled_nulls = df.fillna({"training level":round(avg_training_level),"work experience":round(avg_work_experience)})
display(df_with_filled_nulls)
```
